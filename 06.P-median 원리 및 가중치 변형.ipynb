## 1. 수정된 P-median

- 기본 원리 : 입지후보지와 모든 수요지점 간 거리의 합이 최소가 되는 지점 도출


- 수정된 원리 : 각 수요지점별로 최소거리에 위치한 입지후보지에 할당, 가장 많은 수요지점을 포함하는 지점 도출

앞선 참고 코드에서는 수요지점이 아닌 입지후보지 간의 거리행렬을 생성함

따라서 수요지점과 입지후보지 간의 거리행렬로 변환한 후 참고 코드를 수정하고자 하였으나, 오류 및 시간상의 이유로 보류

**P-median의 기본 원리를 일부 수정함으로써 본 프로젝트에 맞게 변형함**

```python
import numpy as np
import pandas as pd

p = 3     # 선정할 입지 수
행정동 = '석관동'
행정동리스트 = ['석관동','월곡2동','회기동','이문1동','이문2동','묵2동','공릉1동',\
                '월계3동','월계1동','장위1동','장위2동','장위3동']

입지후보지 = pd.read_csv('./data/입지선정후보군_좌표추가.csv', encoding='EUC-KR')
입지후보지 = 입지후보지[입지후보지['행정동'] == 행정동]
후보지_points = np.array([list(i) for i in zip(입지후보지['x좌표'], 입지후보지['y좌표'])])

버스 = pd.read_csv('./data/서울시_버스정류장_행정동추가.csv')
버스 = 버스[버스['행정동'].isin(행정동리스트)]
버스_points = np.array([list(i) for i in zip(버스['X좌표'], 버스['Y좌표'])])
```

```python
from haversine import haversine

havers = []
for i in 후보지_points:
    site = []
    for j in 버스_points:
        site.append(haversine(i,j))
    havers.append(site)
havers
```

```python
# 입지후보지와 버스정류장 간 거리행렬 생성
location = list(입지후보지['시설명'])
location2 = list(버스['정류소명'])

havers_D = dict(zip(location, [dict(zip(location2, i)) for i in havers]))
havers_D
```

### v1. 가중치가 없는 경우

```python
data = pd.DataFrame(havers_D)
print(data.shape)
data.reset_index(inplace=True)
data
```

```python
l1 = []
for i in data.columns:
    for j in range(len(data)):
        if data[i][j] == data.min(axis=1)[j]:
            l1.append(i)
```

```python
for i in data.columns.drop('index'):
    print(i, '=', l1.count(i))
```

```python
전체w = points.shape[0]
버스w = 버스.shape[0]
지하철w = 지하철.shape[0]
주차장w = 주차장.shape[0]
주택w = 주택.shape[0]

m1 = (전체w-버스w)/전체w
m2 = (전체w-지하철w)/전체w
m3 = (전체w-주차장w)/전체w
m4 = (전체w-주택w)/전체w
```

```python
idx = D.index
col = D.columns
lidx = len(idx)
lcol = len(col)

for i in range(lidx):
    for j in range(lcol):
        if D.loc[idx[i], col[j]] == 최소값[i]:
            D.loc[idx[i], col[j]] = m1     # 버스 가중치만 추가
        else:
            D.loc[idx[i], col[j]] = 0
D.sum()
```
