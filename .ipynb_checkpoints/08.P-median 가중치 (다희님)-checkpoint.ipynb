```python
import numpy as np
import pandas as pd
```

```python
# 행정동 = '길동'
# 행정동리스트 = ['길동', '천호3동', '천호1동', '명일1동', '명일2동', '상일동', '둔촌2동', '성내3동', '성내2동']
```

```python
행정동 = '구의제1동'
행정동리스트 = ['구의1동', '화양동','자양1동','자양2동','광장동','중곡4동','구의2동','구의3동', '능동']
```

```python
# 행정동 = '석관동'
# 행정동리스트 = ['석관동','월곡2동','회기동','이문1동','이문2동','묵2동','공릉1동', '월계3동','월계1동','장위1동','장위2동','장위3동']
```

```python
입지후보지 = pd.read_csv('입지후보군_행정동수정.csv', encoding='EUC-KR')
입지후보지 = 입지후보지[입지후보지['행정동'] == 행정동]
후보지_points = np.array([list(i) for i in zip(입지후보지['x좌표'], 입지후보지['y좌표'])])

버스 = pd.read_csv('서울시_버스정류장_행정동추가.csv')
버스 = 버스[버스['행정동'].isin(행정동리스트)]
버스_points = np.array([list(i) for i in zip(버스['X좌표'], 버스['Y좌표'])])

지하철 = pd.read_csv('서울시_지하철역_행정동추가.csv', encoding='EUC-KR', index_col=0)
지하철 = 지하철[지하철['행정동'].isin(행정동리스트)]
지하철_points = np.array([list(i) for i in zip(지하철['경도'], 지하철['위도'])])

주차장 = pd.read_csv('서울시_주차장_행정동추가.csv', encoding='EUC-KR', index_col=0).drop('Unnamed: 0.1', axis=1)
주차장 = 주차장[주차장['행정동'].isin(행정동리스트)]
주차장_points = np.array([list(i) for i in zip(주차장['경도'], 주차장['위도'])])

주택 = pd.read_csv('서울시주택_행정동추가.csv', encoding='EUC-KR', index_col=0).drop('Unnamed: 0.1', axis=1)
주택 = 주택[주택['행정동'].isin(행정동리스트)]
주택_points = np.array([list(i) for i in zip(주택['위도'], 주택['경도'])])   # 위도 경도 바뀜
```

```python
X = list(버스['X좌표']) + list(지하철['경도']) + list(주차장['경도']) + list(주택['위도'])     # 주택 위도 경도 이름 바뀜
Y = list(버스['Y좌표']) + list(지하철['위도']) + list(주차장['위도']) + list(주택['경도'])
```

```python
points = np.array([list(i) for i in zip(X, Y)])
print(points.shape)
points[:2]
```

```python
전체w = points.shape[0]
버스w = 버스.shape[0]
지하철w = 지하철.shape[0]
주차장w = 주차장.shape[0]
주택w = 주택.shape[0]
```

```python
m1 = (전체w-버스w)/전체w
m2 = (전체w-지하철w)/전체w
m3 = (전체w-주차장w)/전체w
m4 = (전체w-주택w)/전체w
```

```python
from haversine import haversine

havers = []
for i in 후보지_points:
    site = []
    for j in 버스_points:
        site.append(haversine(i,j))
    havers.append(site)
```

# 버스

```python
from pulp import *

# 후보지와 버스정류장 간 거리행렬 생성
location = list(입지후보지['시설명'])
location2 = list(버스['정류소명'])

havers_D = dict(zip(location, [dict(zip(location2, i)) for i in havers]))
D = havers_D
```

```python
#col = list(D.keys())
D = pd.DataFrame(D)
D.head()
```

```python
# #col = D.columns
최소값 = list(D.min(axis=1))
```

```python
# 버스
idx = D.index
col = D.columns
lidx = len(idx)
lcol = len(col)

for i in range(lidx):
    for j in range(lcol):
        if D.loc[idx[i], col[j]] == 최소값[i]:
            D.loc[idx[i], col[j]] = m1
        else:
            D.loc[idx[i], col[j]] = 0
```

```python
D.sum()
```
